```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# 1. 계층적 군집분석

```python
X = np.array([[5,3],
    [10,15],
    [15,12],
    [24,10],
    [30,30],
    [85,70],
    [71,80],
    [60,78],
    [70,55],
    [80,91],])
```

```python
labels = range(1, 11)
plt.figure(figsize = (10, 7))
# plt.subplots_adjust(bottom = 0.1)
plt.scatter(X[:, 0], X[:, 1], label = 'True Position')

for label, x, y in zip(labels, X[:, 0], X[:, 1]):
    plt.annotate(label, xy = (x, y), xytext = (-3, 3),
                 textcoords = 'offset points', ha = 'right', va = 'bottom')

# ha : horizontal alignment
# va : vetical alignment
# xytext : 텍스트 위치 
    
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
linked = linkage(X, 'single')
```

```python
labelList = range(1, 11)

plt.figure(figsize = (10, 7))
dendrogram(linked, orientation = 'top', labels = labelList,
           distance_sort = 'descending', show_leaf_counts = True)
plt.show()
```

# 2. 쇼핑 데이터 활용

```python
customer_data = pd.read_csv('./data/shopping-data.csv')
```

```python
customer_data.head()
```

```python
data = customer_data.iloc[:, 3:5].values
```

```python
plt.figure(figsize = (10, 7))
plt.title('Customer Dendrograms')
dend = dendrogram(linkage(data, method = 'ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
cluster.fit_predict(data)
```

```python
plt.figure(figsize = (10, 7))
plt.scatter(data[:, 0], data[:, 1], c = cluster.labels_, cmap = 'rainbow')
```

# 3. boston house 데이터

```python
data = pd.read_csv('./data/Boston_house.csv')
```

```python
target = data['Target']
data = data.drop('Target', axis = 1)
```

```python
data.shape
```

```python
plt.figure(figsize = (10, 7))
plt.title('Boston House Dendrograms')
dend = dendrogram(linkage(data, method = 'ward'))
```

```python
cluster = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward')
cluster.fit_predict(data)
```

```python
np.mean([x for x, y in zip(target, cluster.fit_predict(data)) if y == 0])
```
